# 問題與討論
 #### 分析的資料過於龐大
 a)爬下所有股票資料會占用一定的空間，但爬下來後接下來只需補往後每一天的資料即可
 
 b)進行資料分析時需要花一些時間  ->將資料分割成不同的區間，一方面可以避免一次跑過於龐大的資料量，
   一方面可以讓使用者根據自身需求去做選擇
